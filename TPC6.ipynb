#### __Resumo__: Em primeiro lugar foram importadas funções usadas num programa anterior para a criação de polinómios e tratamento dos mesmos.
#### De seguida, depois do importe destas funções, estas foram usadas para criar um polinómio (o pedido pelo utilizador) e depois mostrar o gráfico do mesmo

```python
import polinomio as pol
```

```python
def polinomio():
    termos=int(input("Quer um polinómio com quantos termos: "))
    poli=pol.criarPolinomio()
    while termos>0:
            c=int(input("Diga qual o coeficiente do Termo:"))
            e=int(input("Diga qual o expoente do Termo: "))
            termo=pol.criarTermo(c,e)
            pol.inserirTermo(poli,termo)
            termos-=1
    return(poli)  

```

```python

import matplotlib.pyplot as plt

p1=polinomio()
x=[n for n in range(-100,100)]
y=[]
for n in x:
    y.append(pol.calcPolinomio(p1,n))
    
plt.plot(x, y)

eixoXx=[x[0]-1,x[-1]+1]
eixoXy=[0,0]
plt.plot(eixoXx,eixoXy, color='red', linewidth = 1) #Desenhar os eixos

plt.xlabel('Abcissas')
plt.ylabel('Ordenadas')
 
plt.title('Gráfico do Polinómio')

plt.show()
```
